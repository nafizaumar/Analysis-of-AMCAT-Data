```python
df.drop_duplicates(inplace=True)  #drop duplicates
```

```python
# Convert relevant columns to datetime (e.g., 'DOJ' for Date of Joining)
df['DOJ'] = pd.to_datetime(df['DOJ'], errors='coerce')
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')
df['DOL'] = pd.to_datetime(df['DOL'], errors='coerce')
```

```python
# Recheck the data types after conversion
df.info()
```

```python
# Univariate Analysis

import matplotlib.pyplot as plt
import seaborn as sns

# Plot histograms for numerical columns
columns_to_plot = ['Salary', '10percentage', '12percentage', 'collegeGPA', 
                   'English', 'Logical', 'Quant', 'ComputerProgramming']

df[columns_to_plot].hist(figsize=(10, 8), bins=30, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
# List of numerical columns
columns_to_plot = ['Salary', '10percentage', '12percentage', 'collegeGPA', 
                   'English', 'Logical', 'Quant', 'ComputerProgramming']

# Plot each numerical column separately as a box plot
for column in columns_to_plot:
    plt.figure(figsize=(10,8))
    sns.boxplot(data=df, y=column)
    plt.title(f'Box Plot for {column}')
    plt.ylabel(column)
    plt.show()
```

```python
# Display only the top 10 categories
categorical_columns = ['Designation', 'JobCity', 'Gender']

for column in categorical_columns:
    plt.figure(figsize=(12, 8))
    top_values = df[column].value_counts().nlargest(20)
    sns.countplot(data=df[df[column].isin(top_values.index)], x=column, order=top_values.index)
    plt.title(f"Top 10 Count Plot for {column}")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
# Get the unique values and their counts in the Degree column
degree_counts = df['Degree'].value_counts()
print(degree_counts)
```

```python
# Pie chart for degree distribution (top 4 degrees shown)

plt.figure(figsize=(8, 8))

# Explode for better visibility of smaller slices
explode = [0.1 if value < 50 else 0 for value in degree_counts.values]

degree_counts.plot.pie(autopct='%1.1f%%', startangle=90, labeldistance=1.1, explode=explode,
                       colors=sns.color_palette('Set3'))

plt.title('Proportion of Degrees')
plt.ylabel('')  # Hide y-label for a cleaner look
plt.tight_layout()
plt.show()
```

```python
# Get the top 10 most common specializations
top_specializations = df['Specialization'].value_counts().nlargest(10)

# Plot a horizontal bar chart for the top 10 specializations
plt.figure(figsize=(12, 8))
sns.barplot(x=top_specializations.values, y=top_specializations.index, palette='coolwarm')
plt.title('Top 10 Specializations')
plt.xlabel('Count')
plt.ylabel('Specialization')
plt.tight_layout()
plt.show()
```

```python
#Bivariate Variables:
#1. Numerical Vs Numerical Variables: Scatter Plot & Correlation
    
# Scatter plot to explore the relationship between Salary and College GPA
plt.figure(figsize=(10, 6))
sns.scatterplot(x='collegeGPA', y='Salary', data=df)
plt.title('Scatter Plot of Salary vs College GPA')
plt.xlabel('College GPA')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation between Salary and College GPA
correlation = df[['collegeGPA', 'Salary']].corr()
print(correlation)
```

```python
# 2. Categorical Vs Numerical Variables: Box Plot & Bar Plot

# Get the top 10 specializations by count
top_10_specializations = df['Specialization'].value_counts().nlargest(10).index

# Filter the DataFrame to include only the top 10 specializations
df_top_specializations = df[df['Specialization'].isin(top_10_specializations)]

# Box plot to visualize the distribution of Salary across the top 10 Specializations
plt.figure(figsize=(18, 12))
sns.boxplot(x='Specialization', y='Salary', data=df_top_specializations)
plt.title('Box Plot of Salary by Top 10 Specializations')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Bar plot showing average salary for each Specialization
plt.figure(figsize=(12, 8))
sns.barplot(x='Specialization', y='Salary', data=df, estimator=np.mean)
plt.title('Average Salary by Specialization')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
#Categorical Vs Categorical Variables: Heatmap

# Get the top 10 specializations by count
top_10_specializations = df['Specialization'].value_counts().nlargest(10).index

# Filter the DataFrame to include only the top 10 specializations
df_top_specializations = df[df['Specialization'].isin(top_10_specializations)]

# Create a cross-tabulation of Specialization and Gender for the top 10 specializations
cross_tab = pd.crosstab(df_top_specializations['Specialization'], df_top_specializations['Gender'])

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='coolwarm')
plt.title('Heatmap of Top 10 Specializations vs Gender')
plt.tight_layout()
plt.show()
```

```python
#4. Multiple Numerical Variables: Pair Plot
# Select numerical columns for the pair plot
numerical_columns = ['Salary', 'collegeGPA', '10percentage', '12percentage']

# Create a pair plot
sns.pairplot(df[numerical_columns])
plt.suptitle('Pair Plot of Numerical Columns', y=1.02)
plt.show()
```
